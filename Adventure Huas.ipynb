```python
import pandas as pd
import numpy as np
import datetime
import seaborn
import matplotlib as plt
```

```python
# Other
aquisition = pd.read_csv('data/Adventure_Huas/Google_Analytics/aquisition.csv',skiprows=[i for i in range(0,5)])
ecom = pd.read_csv('data/Adventure_Huas/Google_Analytics/ecommerce.csv',skiprows=[i for i in range(0,6)])
users = pd.read_csv('data/Adventure_Huas/Google_Analytics/users.csv',skiprows=[i for i in range(0,5)])

aquisition.reset_index(inplace=True)
ecom.reset_index(inplace=True)
users.reset_index(inplace=True)

# Behavior
behavior_first = pd.read_csv('data/Adventure_Huas/Google_Analytics/behavior_1_5000.csv',skiprows=[i for i in range(0,5)])
behavior_middle = pd.read_csv('data/Adventure_Huas/Google_Analytics/behavior_5001_10000.csv',skiprows=[i for i in range(0,5)])
behavior_last = pd.read_csv('data/Adventure_Huas/Google_Analytics/behavior_10001_14300.csv',skiprows=[i for i in range(0,5)])

# Combine Behaviors
behavior_total = pd.concat([behavior_first, behavior_middle, behavior_last])
behavior_total = behavior_total.reset_index(drop=True)
behavior_total.reset_index(inplace=True)
```

```python
# This section will take each dataframe and find nulls, datatypes, summary stats, and column counts

variables = [aquisition, ecom, users, behavior_total]
names = ['aquisition', 'ecom', 'users', 'behavior_total']

b = 0

for i in variables:
    print(names[b])
    print('------------------------------')
    print('***************** Nulls *****************')
    print(i.isnull().sum())
    print('***************** Data Types *****************')
    print(i.dtypes)
    print('***************** Summary Stats *****************')
    print(i.describe(include='all'))
    print('***************** Column Stats *****************')
    for col in i:
        if col != 'index':
            print(col)
            print(i[['index',col]].groupby(col).count())
    print('------------------------------')
    b += 1
```
